## Cleaning

```python
import nltk
nltk.download ('stopwords')
nltk.download ('punkt')
```

```python
##Membuat kalimat yang menggunakan huruf kapital menjadi huruf kecil
kalimat = "Saya tidak mau! Pokoknya hari ini kamu harus pulang!!!."
lower_case = kalimat.lower()
print(lower_case)
```

```python
##Menghapus numerik pada kalimat
import re
kalimat = "Halo! Perkenalkan aku Budi. Umurku 20 tahun. Salam kenal!."
hasil = re.sub(r"\d+", "", kalimat)
print(hasil)
```

```python
##Mengubah numerik menjadi bentuk karakter
import inflect
kalimat = "There are 3 books that I haven't read, 2 of them are still wrapped"
p = inflect.engine()
hasil = []
for i in kalimat.split():
  if i.isnumeric():
    i = p.number_to_words(i)
  hasil.append(i)
hasil = " ".join(hasil)
print(hasil)
```

```python
##Penghapusan non karakter
import string
kalimat = "Selamat berakhir pekan teman-teman semua :)))"
hasil = kalimat.translate(str.maketrans("","",string.punctuation))
print(hasil)
```

```python
##Menghapus URL pada kalimat
kalimat = "Tempat ini kotor sekali! Aku tidak mau kesini lagi! http://www.tempatkotor.com"
print(re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", kalimat))
```

```python
kalimat = "Sepertinya aku belum coba ini kamu gimana? @kawanku #teman"
kalimat1 = "Lelah banget pulang sekolah @jakasetia2. Mau tidur aja rasanya #capek #lelah #letih"
print(re.sub("#[A-Za-z0-9]+","",kalimat)) #tanpa hashtag
print(re.sub("@[A-Za-z0-9]+","",kalimat1)) #tanpa mention
```

## Tokenizing

```python
##Memisahkan kalimat berdasarkan spasi 
kalimat = "Hebat! Kamu memang pantas jadi juara!!"
pisah = kalimat.split()
print(pisah)
```

```python
##Memisahkan kalimat berdasarkan spasi dan tanda baca
from nltk.tokenize import word_tokenize 
 
kalimat = "Kenapa kamu masih mencari yang lain????" 
tokens = nltk.tokenize.word_tokenize(kalimat)
print(tokens)
```

```python
##Memisahkan 2 kalimat berdasarkan spasi setelah tanda baca
from nltk.tokenize import sent_tokenize
kalimat = "Ada 3 buku yang belum aku baca 2 diantaranya masih dibungkus. Masakan mama enak sekali!."
 
tokens = nltk.tokenize.sent_tokenize(kalimat)
print(tokens)
```

```python
from nltk.corpus import stopwords
 
kalimat = "Halo! Perkenalkan aku Budi. Umurku 20 tahun. Salam kenal!. Ada 3 buku yang belum aku baca 2 diantaranya masih dibungkus."
kalimat = kalimat.translate(str.maketrans('','',string.punctuation)).lower()
 
tokens = word_tokenize(kalimat)
listStopword =  set(stopwords.words('indonesian'))
 
removed = []
for t in tokens:
    if t not in listStopword:
        removed.append(t)
 
print(removed)
```
