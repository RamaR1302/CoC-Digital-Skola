##4. Create a new column named “age_interval”

```python
df.loc[ (df.Age < 25), 'Age_Interval'] = '< 25'
df.loc[ (df.Age >= 25) & (df.Age <= 35 ), 'Age_Interval'] = '25 - 35'
df.loc[ (df.Age > 35) & (df.Age < 51 ), 'Age_Interval'] = '36 - 50'
df.loc[ (df.Age > 50), 'Age_Interval'] = '> 50'

df
```

##5. After creating the new column in question number 4, which age interval has the highest percentage of Survived ?

```python
Percent_Survived = df.groupby('Age_Interval').sum()
Percent_Survived
```
